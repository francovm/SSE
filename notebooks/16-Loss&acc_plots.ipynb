```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pylab import rcParams
```

# Autoencoder

```python
# load  dataset
df = pd.read_csv('/home/francovm/Projects/SSE/data/Visualization/Data/Autoencoder-classifier_LSTM_SSE_20window-stand-2.csv', sep=';', encoding='utf-8')
```

```python
# summarize history for accuracy
plt.plot(df['accuracy'].values,linewidth=5.0,color='maroon')
plt.plot(df['val_accuracy'].values,linewidth=5.0,color='gray')
plt.title('Model accuracy',fontsize=30, ha="center",pad=20)
plt.ylabel('Accuracy',fontsize=30,labelpad=20)
plt.xlabel('Epoch',fontsize=30)
plt.legend(['training data', 'Validation data'], loc='upper left',fontsize=30)

# Get current size
fig_size = plt.rcParams["figure.figsize"]
#  Set figure width to 18 and height to 4
fig_size[0] = 30
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

ax = plt.subplot()

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
# set the grid on
# ax.grid('on')

plt.rc('grid', linestyle="-", color='black',alpha='0.3')
plt.grid(True)

ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 30)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 30)

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
# plt.axhline(y=0.952, color='r', linestyle='-')

plt.show()
```

```python
# summarize history for accuracy
plt.plot(df['loss'].values,linewidth=5.0,color='maroon')
plt.plot(df['val_loss'].values,linewidth=5.0,color='gray')
plt.title('Model Loss',fontsize=30, ha="center",pad=20)
plt.ylabel('Loss',fontsize=30,labelpad=20)
plt.xlabel('Epoch',fontsize=30)
plt.legend(['training data', 'Validation data'], loc='upper left',fontsize=30)

# Get current size
fig_size = plt.rcParams["figure.figsize"]
#  Set figure width to 18 and height to 4
fig_size[0] = 30
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

ax = plt.subplot()

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
# set the grid on
plt.rc('grid', linestyle="-", color='black',alpha='0.3')
plt.grid(True)

ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 30)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 30)

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
# plt.axhline(y=0.952, color='r', linestyle='-')

plt.show()
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.plot(df['loss'].values,linewidth=3.0,color='maroon')
ax1.plot(df['val_loss'].values,linewidth=3.0,color='gray')
ax2.plot(df['accuracy'].values,linewidth=3.0,color='maroon')
ax2.plot(df['val_accuracy'].values,linewidth=3.0,color='gray')


ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data')
ax2.set_ylabel('Y2 data')

plt.show()
```
